## top 10 most rated books

```python
df.groupby('Book Name')['Book_ratings_count'].sum().sort_values(ascending=False).head(10)
```

## Top 10 books with highest “Book_average_rating”

```python
df.groupby('Book Name')['Book_average_rating'].mean().sort_values(ascending=False).head(10)
```

```python
df['Author_Rating'].value_counts()
```

```python
# Keep only realistic publishing years
df_clean = df[(df['Publishing Year'] >= 1800) & (df['Publishing Year'] <= 2025)]

plt.hist(df_clean['Publishing Year'],bins = 20)
plt.title("Distribution of Publishing Years")
plt.xlabel("Publishing Year")
plt.ylabel("Frequency")


plt.tight_layout()
plt.show()

```

```python
df['genre'].value_counts().plot(kind='bar')
plt.xlabel('Genre')
plt.ylabel('Number of Books')
plt.title('Number of books in each Genre')
plt.show()
```

```python
df['Author_Rating'].value_counts().plot(kind='pie',autopct='%.2f%%',ylabel='',title='Author Rating Distribution')
```

```python
df['Publisher '].value_counts().plot(kind='bar',xlabel='Publisher',
    ylabel='Number of Books',
    title='Publishers by Number of Books')
```

```python
print(df.columns)
```

```python
df.groupby('genre')['publisher revenue'].sum().plot(kind='bar',xlabel='Genre',
    ylabel='Total Publisher Revenue',
    title='Total Publisher Revenue per Genre')
```

```python
df.groupby('Book Name')['units sold'].sum().sort_values(ascending=False).head(10).plot(kind='bar',xlabel='Book Name',
    ylabel='Unit Sold',
    title='Top 10 Books by Sales')
```

```python
df.groupby('genre')['publisher revenue'].sum().plot(kind = 'bar',title = 'Revenue by genre', xlabel = 'Genre', ylabel = 'Revenue')
```

```python
df.groupby('Publisher ')['gross sales'].sum().plot(kind='bar',xlabel='Publisher',ylabel='Gross Sales',title='Publisher vs Gross Sales')
```
